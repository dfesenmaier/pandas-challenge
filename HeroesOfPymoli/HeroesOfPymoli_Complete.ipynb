## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
grouped_df = df.groupby(["Gender"])
unique_df = grouped_df.nunique()

total_gender = unique_df["SN"].sum()

count = unique_df["SN"].unique()
percentage = unique_df["SN"]/ total_gender

gender_summary = pd.DataFrame({
        
        "Count":count,
        "Percentage of Players":percentage
})

gender_summary["Percentage of Players"] = gender_summary["Percentage of Players"].map("{:,.2%}".format) 
gender_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.90, 19.9, 29.9, 39.9, 49.9]
group_names = ["<10", "10-19", "20-29", "30-39", "40+"]

df["Age Group"] = pd.cut(df["Age"],bins, labels=group_names)

age_groupes = df.groupby("Age Group")

total_count_age = age_groupes["SN"].nunique()

percentage_by_age = (total_count_age/total_players)

age_demographics = pd.DataFrame({
    
    "Total Count": total_count_age,
    "Age Group Percentage": percentage_by_age
})

age_demographics["Age Group Percentage"] = age_demographics["Age Group Percentage"].map("{:,.2%}".format) 

age_demographics.sort_index(axis=0)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
purchase_count_age = age_groupes["Purchase ID"].count()

avg_purchase_price_age = age_groupes["Price"].mean()

total_purchase_value = age_groupes["Price"].sum()

avg_purchase_per_person_age = total_purchase_value/total_count_age

age_demographics = pd.DataFrame({"Purchase Count": purchase_count_age,
                                 "Average Purchase Price": avg_purchase_price_age,
                                 "Average Purchase Total/Person": avg_purchase_per_person_age})
age_demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_stats = df.groupby("SN")

total_purchases = spender_stats["Purchase ID"].count()
average_purchases = spender_stats["Price"].mean()
total_spender = spender_stats["Price"].sum()

top_spenders_df = pd.DataFrame({
    
    "Purchase Count" : total_purchases,
    "Average Purchase Price" : average_purchases,
    "Total Purchase Value" : total_spender
    
})

top_spenders_final_df = top_spenders_df.sort_values(["Total Purchase Value"], ascending=False).head()

top_spenders_final_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items_df = df[["Item ID", "Item Name", "Price"]]
item_stats = items_df.groupby(["Item Name", "Item ID"])
item_counts = item_stats["Price"].count()
item_values = item_stats["Price"].sum()

item_price = item_values/item_counts

popular_items = pd.DataFrame({
    
    "Purchase Count" : item_counts,
    "Item Price" : item_price,
    "Total Purchase Value" : item_values
})
top_items_final_df = popular_items.sort_values(["Purchase Count"], ascending=False).head()

top_items_final_df
```
